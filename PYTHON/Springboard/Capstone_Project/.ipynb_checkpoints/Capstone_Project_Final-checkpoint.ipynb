CAN PERSONALITY FEATURES PREDICT SUBSTANCE ABUSE ?
==================




The purpose of this notebook is to explore raw data collected by a group of psychologists and mathematicians from Cornell University (https://arxiv.org/abs/1506.06297).  The group's purpose for this study was to examine possible correlation between 18 central nervous system psychoactive drugs, 7 personality traits and 5 behavioral/personality features.  1884 participants were part of this study; 941 were female, 943 were male.  Below is a list of the items analyzed:

**A) 12 personality attributes are known for each participant**

    8. Age
    9. Gender 
    10. Education (level of education of participant)
    11. Country
    12. Ethnicity.

**B) 18 central nervous system psychoactive drugs**

**Our objective in this review is to explore and analyze the anonymous online database survey utilized in the above publication and focus on two things:** 

	1. The potential effect between the “Big Five Personality Traits,” 
    demographic data and drug consumption for different psychoactive 
    drugs.  
	
	2. Employment of classification analysis (by optimizing 
    hyperparameters in logistic regression) predict the probability binary correlation of drug consumption for each individual based on 
		
		a. personality attributes vs all psychoactive drugs 
		b. personality attributes vs the opioids cluster (heroin, 
        methadone, and legal highs)

```python
#G.  Drugs(i-xvii)

# A code will be iterated over all drugs in which it will replace the object labels of each drug, for each
# paritcipants answer with either the value "0" or "1".  If an individual HAS ever taken a drug within the last year,
# month, week or day (of when they answered a YES for that question) then the objects answer will be chanaged 
# from CL3, CL4, CL5, CL6,respectovely, to a 1.  If that person has NOT take a drug in last decade, over a decase ago
# or never used, then the objects answer will be changed from CL0, CL1, CL2, respectively, to a 0.  It is important to
# state the reason this was done was to create a binary classification problem in which focus is made for participants 
# who are sensitive to frequent drub abuse.

# Drug Values of original df will be replaces with either a 0 or 1 where 
#             0= never used drug
#             1= Used drug 


drug_usage_dict= {
    "CL0":"Never Used: 0", 
    "CL1":"Used over a Decade Ago (considered not used): 0", 
    "CL2":"Used in Last Decade: 0",  
    "CL3":"Used in Last Year : 1",  
    "CL4":"Used in Last Month: 1",  
    "CL5":"Used in Last Week: 1",  
    "CL6":"Used in Last Day: 1",
}
 

#dictionary
drug_dict = {0: ['CL0', 'CL1','CL2'], 1: ['CL3', 'CL4','CL5','CL6']}

#itiration of dictionary over drug columns 

for k, v in drug_dict.items():
    data2.loc[data2.alcohol.str.upper().isin(v), 'alcohol'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.amphet.isin(v), 'amphet'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.amyl.str.upper().isin(v), 'amyl'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.benzos.str.upper().isin(v), 'benzos'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.caff.str.upper().isin(v), 'caff'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.cannabis.str.upper().isin(v), 'cannabis'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.chocolate.str.upper().isin(v), 'chocolate'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.coke.isin(v), 'coke'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.crack.str.upper().isin(v), 'crack'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.ecstasy.str.upper().isin(v), 'ecstasy'] = k
        
for k, v in drug_dict.items():
    data2.loc[data2.heroin.str.upper().isin(v), 'heroin'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.ketamine.str.upper().isin(v), 'ketamine'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.legalh.str.upper().isin(v), 'legalh'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.LSD.str.upper().isin(v), 'LSD'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.meth.str.upper().isin(v), 'meth'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.mushrooms.str.upper().isin(v), 'mushrooms'] = k    
    
for k, v in drug_dict.items():
    data2.loc[data2.nicotine.str.upper().isin(v), 'nicotine'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.semer.str.upper().isin(v), 'semer'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.VSA.str.upper().isin(v), 'VSA'] = k
    
data2.head(5)
```

#  c. Logistic Regression Model for All Drugs

**Steps for Logistic Regression**

As we build our binary classification model, we do so, with the requirement that it not only 
fits our current dataset well, but also, new unseen data.  To perform this task, 
we have chosen Logistic Regression.  Logistic Regression will split our data into two parts:
a training and test set.  First, we will train or fit on the training set and then test for accuracy on the testing test.  Once the data has gone through the train_test_splitting phase, the model can then be tuned using a hyperparameter classifier via two routes: K-fold cross validation or Grid Search Cross Validation.  Below are the steps.
 

**Tuning model with K-fold cross validation**

1. First we use the cv_score function to perform K-fold cross-validation.  
2. Second, we use the scoring function to test each fold.
3. Third, through the now optimized cv_function, we find find best model C parameters based only on the training set, without showing the model test set at all .
4. Fourth, we create a logistic regression model with the newly calculated  best regularization parameter C and train, only again,  on the training data.
5. Fifth, finally, we train logistic regression (using the best regularization parameter C and the predictive function) on TEST data
6. Sixth, we use the accuracy_score function to test on the models accuracy for predicting y values.


**Tuning model with Scikit-leanr’s Grid Search Cross Validation**

Grid Search CV uses the best model over the training set, in comparison to K-fold Cross validation, which utilizes a iterations of loops over model parameters.

1. First we use the grid search function to apply logistic regression on the parameters using cross validation
2. Second, We fit the training set, not the test data, and find the best 
    1. Estimator C and its respective parameters 
    2. Best score for the test data and confirmation of chosen parameters
        1. mean scores of grid scores
        2. stds
        3. parameters 
3. Third, we use the best regularization estimator C found in 2.A, to fit the testing data 
4. Fourth,  we use the accuracy_score function to test on the  models accuracy for predicting y values.


Note, that for K-fold/Grid Search cross validation, a very important parameter is shared, the regularization parameter C.  Binary classification models have hyperparameters that can be tuned for better performance.  In Logistic Regression, the most important parameter to tune is the regularization parameter C.  This parameter controls for unlikely high regression coefficients and can be used as a method for feature selection when data is sporadically dispersed.  In order to select the best regularization parameter C for our model, we must follow a few steps.

The system creates K train/test splits (into 2 pieces) averages results together .  This allows the model to be trained and tested independently.  It's a better estimate of out-of-shape sample performance than just the test/train splitting, however, it runs K times slower than train/test sp.

#### a. Train set

*Logistic Regression Model Evaluation Metrics using Optimized Regularization Parameter C and K-Fold Cross Validation: TRAINING SET*

    1) K-folds Train/test splitting
    2) Tune model
    3) Set up of Logistic Regression model using best Regularization              Parameter C
    4) Class prediction 
    5) Accuracy 
    6) Null Accuracy
    7) Sensitivity/Recall Metric
    8) Precision Metric
    9) ROC 
    10) AUC (Area under the Curve)
    11) Classification Report

```python
# 1) K-folds Train/test splitting for train data

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

clf2 = LogisticRegression() 
def cv_score_train_k(clf2, x_allk, y_allk, score_func=accuracy_score):
    result = 0
    nfold = 5
    for train, test in KFold(nfold).split(x_allk): # split data into train/test groups, 5 times
        clf.fit(x_allk[train], y_allk[train]) # fit
        result += score_func(clf.predict(x_allk[test]), y_allk[test]) # evaluate score function on held-out data
    return result / nfold # average
```

```python
clf3 = LogisticRegression()
CV_score_train_all_drugs_no_C = cv_score_train_k(clf3, Xlr_all, ylr_all)
cv_score_train_k(clf3, Xlr_all, ylr_all)
print('Cross Validation precision for All drugs train data with k fold without using parameter C: ' + str((CV_score_train_all_drugs_no_C)))
```

```python
# 2) Tune Model for all drugs k-fold

#the grid of parameters to search over
Cs = [0.001, 0.1, 1, 10, 100]
from sklearn.linear_model import LogisticRegression
highest_score = 0

clf4=LogisticRegression()

for C in Cs:
        clf4 = LogisticRegression(C=C)
        avg_score = cv_score_train_k(clf4, Xlr_all, ylr_all)

        if avg_score > highest_score:
            highest_score = avg_score
            C1=C

print('Highest score parameter using the CV_function for All drugs of train data and C: ' + str((highest_score)))
print('')
print('Best Regularization Parameter C: ' + str((C1)))
```

```python
# 3) Set up of Logisitc Regression model using best Regularization Parameter C for k-fold training

clf5=LogisticRegression(C=C1)
clf5.fit(Xlr_all, ylr_all)
Fit_train_all_drugs_yes_CV = clf5.fit(Xlr_all, ylr_all)
print(Fit_train_all_drugs_yes_CV)
```

##   i. K-fold Cross Validation: 
        
   #### a. Train set
   #### b. Test set

*Logistic Regression Model Evaluation of K train/test splitting*

The system creates K train/test splits (into 2 pieces) averages results together .  This allows the model to be trained and tested independently.  It's a better estimate of out-of-shape sample performance than just the test/train splitting, however, it runs K times slower than train/test sp.

#### a. Train set

*Logistic Regression Model Evaluation Metrics using Optimized Regularization Parameter C and K-Fold Cross Validation: TRAINING SET*

    1) K-folds Train/test splitting
    2) Tune model
    3) Set up of Logistic Regression model using best Regularization Parameter C
    4) Class prediction 
    5) Accuracy 
    6) Null Accuracy
    7) Sensitivity/Recall metric 
    8) Precision Metric
    9) ROC 
    10) AUC (Area under the Curve)
    11) Classification Report

```python
# 1) K-folds Train/test splitting for training data

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

clf15 = LogisticRegression()

def cv_score_train_k_o(clf15, x_opi, y_opi, score_func=accuracy_score):
    result = 0
    nfold = 5
    for train, test in KFold(nfold).split(x_opi): # split data into train/test groups, 5 times
        clf15.fit(x_opi[train], y_opi[train]) # fit
        result += score_func(clf15.predict(x_opi[test]), y_opi[test]) # evaluate score function on held-out data
    return result / nfold # average


```

```python
clf16 = LogisticRegression()
CV_score_train_opi_no_C = cv_score_train_k_o(clf16, Xlr_opioids, ylr_opioids)
cv_score_train_k_o(clf16, Xlr_opioids, ylr_opioids)
print('Cross Validation precision for Opioids train data with k fold without using parameter C: ' + str((CV_score_train_opi_no_C)))
```

```python
# 2) Tune Model for Opioids k-fold CV

clf17 = LogisticRegression()
#the grid of parameters to search over

Cs = [0.001, 0.1, 1, 10, 100]
from sklearn.linear_model import LogisticRegression
highest_score = 0

for C in Cs:
        clf17 = LogisticRegression(C=C)
        avg_score = cv_score_train_k_o(clf17, Xlr_opioids, ylr_opioids)

        if avg_score > highest_score:
            highest_score = avg_score
            C3=C
  
print('Highest score parameter using the CV_function for Opioids drugs of train data: ' + str((highest_score)))
print('')
print('Best Regularization Parameter C: ' + str((C3)))
```

```python
# 3) Set up of Logisitc Regression model using best Regularization Parameter C for k-fold training

clf18 =LogisticRegression(C=C3)
clf18.fit(Xlr_opioids, ylr_opioids)
Fit_train_opi_yes_CV = clf18.fit(Xlr_opioids, ylr_opioids)
print(Fit_train_opi_yes_CV)

```
