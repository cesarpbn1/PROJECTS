CAN PERSONALITY FEATTURES PREDICT DRUG ABUSE ?
==================




The purpose of this notebook is to explore raw data collected by a group of psychologists and mathematicians from Cornell Univeristy (https://arxiv.org/abs/1506.06297).  The group's purpose for this study was to examine possible correlation between 18 central nervous system psychoactive drugs, 7 personality traits and 5 pbehavioral/personality features.  1884 participants were part of this study; 941 were female, 943 were male.  Below is a list of the items analyzed:

**A) 12 personality attributes are known for each participant**

    1. Openness_to_experience
    2. Conscientiousness 
    3. Extraversion
    4. Agreeableness
    5. Neuroticism 
    6. Impulsiveness
    7. Sensation Seeking (S.S)
    8. Age
    9. Gender 
    10. Education (level of education of participant)
    11. Country
    12. Ethnicity.

**B) 18 central nervous system psychoactive drugs**
 
    1. Alcohol
    2. Amphetamines Amyl nitrite
    3. Amyl nitrite 
    4. Benzodiazepines
    5. Caffeine
    6. Chocolate
    7. Cocaine powder (coke)
    8. Cocaine solid (crack)
    9. Ecstasy
    10. Heroin 
    11. Ketamine
    12. Legal highs
    13. LSD 
    14. Methadone
    15. Magic mushrooms 
    16. Nicotine
    17. Semeron (Fictitious and introduced to identify over-claimers)
    18. VSA   (Violent solvent abuse)




**Our objective in this review is to explore and analyze the anonymous online database survey utilized in the above publication and focus on two things:** 

	1. The potential effect between the “Big Five Personality Traits,” 
    demographic data and drug consumption for different psychoactive 
    drugs.  
	
	2. Employment of classification analysis (by optimizing 
    hyperparameters in logistic regression) predict the probability binary correlation of drug consumption for each individual based on 
		
		a. personality attributes vs all psychoactive drugs 
		b. personality attributes vs the opioids cluster (heroin, 
        methadone, and legal highs)

```python
#G.  Drugs(i-xvii)

# A code will be iterated over all drugs in which it will replace the object labels of each drug, for each
# paritcipants answer with either the value "0" or "1".  If an individual HAS ever taken a drug within the last year,
# month, week or day (of when they answered a YES for that question) then the objects answer will be chanaged 
# from CL3, CL4, CL5, CL6,respectovely, to a 1.  If that person has NOT take a drug in last decade, over a decase ago
# or never used, then the objects answer will be changed from CL0, CL1, CL2, respectively, to a 0.  It is important to
# state the reason this was done was to create a binary classification problem in which focus is made for participants 
# who are sensitive to frequent drub abuse.

# Drug Values of original df will be replaces with either a 0 or 1 where 
#             0= never used drug
#             1= Used drug 


drug_usage_dict= {
    "CL0":"Never Used: 0", 
    "CL1":"Used over a Decade Ago (considered not used): 0", 
    "CL2":"Used in Last Decade: 0",  
    "CL3":"Used in Last Year : 1",  
    "CL4":"Used in Last Month: 1",  
    "CL5":"Used in Last Week: 1",  
    "CL6":"Used in Last Day: 1",
}
 

#dictionary
drug_dict = {0: ['CL0', 'CL1','CL2'], 1: ['CL3', 'CL4','CL5','CL6']}

#itiration of dictionary over drug columns 

for k, v in drug_dict.items():
    data2.loc[data2.alcohol.str.upper().isin(v), 'alcohol'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.amphet.isin(v), 'amphet'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.amyl.str.upper().isin(v), 'amyl'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.benzos.str.upper().isin(v), 'benzos'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.caff.str.upper().isin(v), 'caff'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.cannabis.str.upper().isin(v), 'cannabis'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.chocolate.str.upper().isin(v), 'chocolate'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.coke.isin(v), 'coke'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.crack.str.upper().isin(v), 'crack'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.ecstasy.str.upper().isin(v), 'ecstasy'] = k
        
for k, v in drug_dict.items():
    data2.loc[data2.heroin.str.upper().isin(v), 'heroin'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.ketamine.str.upper().isin(v), 'ketamine'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.legalh.str.upper().isin(v), 'legalh'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.LSD.str.upper().isin(v), 'LSD'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.meth.str.upper().isin(v), 'meth'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.mushrooms.str.upper().isin(v), 'mushrooms'] = k    
    
for k, v in drug_dict.items():
    data2.loc[data2.nicotine.str.upper().isin(v), 'nicotine'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.semer.str.upper().isin(v), 'semer'] = k
    
for k, v in drug_dict.items():
    data2.loc[data2.VSA.str.upper().isin(v), 'VSA'] = k
    
data2.head(5)
```
