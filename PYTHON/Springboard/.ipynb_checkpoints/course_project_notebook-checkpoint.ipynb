CAN PERSONALITY TRAITS PREDICT DRUG ABUSE ?
==================




The purpose of this notebook is to explore raw data collected by a group of psychologists and mathematicians from Cornell Univeristy (https://arxiv.org/abs/1506.06297).  The group's purpose for this study was to examine possible correlation between 18 central nervous system psychoactive drugs, 7 personality traits and 5 personality features.  1884 participants were part of this study; 941 were female, 943 were male.  Below is a list of the items analyzed:

**A) 12 personality attributes are known for each participant**

    1. Openness_to_experience
    2. Conscientiousness 
    3. Extraversion
    4. Agreeableness
    5. Neuroticism 
    6. Impulsiveness
    7. Sensation Seeking (S.S)
    8. Age
    9. Gender 
    10. Education (level of education of participant)
    11. Country
    12. Ethnicity.

**B) 18 central nervous system psychoactive drugs**
 
    1. Alcohol
    2. Amphetamines Amyl nitrite
    3. Amyl nitrite 
    4. Benzodiazepines
    5. Caffeine
    6. Chocolate
    7. Cocaine powder (coke)
    8. Cocaine solid (crack)
    9. Ecstasy
    10. Heroin 
    11. Ketamine
    12. Legal highs
    13. LSD 
    14. Methadone
    15. Magic mushrooms 
    16. Nicotine
    17. Semeron (Fictitious and introduced to identify over-claimers)
    18. VSA   (Violent solvent abuse)




Our objective for this project is to explore and analyze the data.  Through employment of logistic classification regression and optimization of hyperparameter classifiers, we are to report possible co-existence between the 12 personality attributes for each participant and their respective susceptibibilty to consumption of all listed drugs and the speficic opioids cluster (heroin, methadone, and legal heroin)

```python
# 1.3) Personality Weighted Conversion

# The uploaded dictionary values in 1.i. must be converted to a weighted percentage.  Their current values are answers 
# to questions in  specifically designed surveys for each personalty trait and in its current state, are in raw form. 
# The coonversion equation is listed below.  The surveys were independently conducted by another research group. 

# Equation: Percentage of Weighted Personality of Participant = (Weighter value/total population) *100

population = 1884
data2['neuroticism'] =(data2.neuroticism/population)*100
data2['extraversion'] =(data2.extraversion/population)*100
data2['openness_to_experience'] =(data2.openness_to_experience/population)*100
data2['agreeableness'] =(data2.agreeableness/population)*100
data2['conscientiousness'] =(data2.conscientiousness/population)*100
data2['impulsivity'] =(data2.impulsivity/population)*100
data2['SS'] =(data2.SS/population)*100
data2.head(5)
```

```python
# GENDER

# We will explore if a correlation exists between gender and opioids using several statisitcal methods. 

# lets import packages to be used 
from scipy import stats
from scipy.stats import norm


#lets examine genders 

gender_1= data3.groupby('gender').size()
print(gender_1)
```

```python
# A quick analysis of one the NEOAC traits, neuroticism.  Since all the personalties share a resemblence to one another,
# it's adequate to use examine one and make an overview analysis

print(stats.normaltest(data3.heroin))
print(len(data3.neuroticism)) # nr of samples 
sns.distplot(data3['neuroticism'], fit=norm)
print(data3.neuroticism.std())

#As we can see this, the p value is incredibly small and something to consider when exploring the hypothesis test.  
#Our population, n= 1884, has n>30, and per our group, the curve resembles that of a binomial distribution which is
# important because that offers candidacy for machine learning categorical exploration.
```

```python
#Question 1: What test is appropriate for this problem? Does CLT apply?
#lets examine and define how many characters are in the gender category 

data3["neuroticism"] = data3[["neuroticism"]].astype(int)
df_male = data3[data3.gender=='male']
df_female = data3[data3.gender=='female']
num_male = len(df_male)
num_female = len(df_female)
print('Total population: ' + str(len(data3.gender))) 
print('Total Male participants: ' + str(num_male))
print('Total Female participants: ' + str(num_female))

#The first thing we look for to determine if the distribution is normal it to examine if the total
# population (n) is >30. Based on previous cell, our n = 1884, 943 for male, 941 for female.  Secondly, 
# we need to determine what type of problem this is.  It is a binomial problem as random 
# candidates either took a drug or did not take a drug.  Due to the size of samples, a binomial distribution can be 
# approximated by a normal distribution per the CLT.
```

```python
#Question 2: Can we createa a null and alternate hypotheses?

#H1 : prob_m = prob_f
#H0 : prob_m != prob_f

#The null hypothesis or H0 is that there is no difference in the mean rates between someone taking a drug or not taking
# a drug.  The alternate hypothesis or H1 is that there is a difference between someone taking a drug and someone 
# not taking a drug
```

#  c. Logistic Regression Model for All Drugs

**Steps for Logistic Regression**

As we build our binary classification model, we do so, with the requirement that it not only 
fits our current dataset well, but also, new unseen data.  To perform this task, 
we have chosen Logistic Regression.  Logistic Regression will split our data into two parts:
a training and test set.  First, we will train or fit on the training set and then test for accuracy on the testing test.  Once the data has gone through the train_test_splitting phase, the model can then be tuned using a hyperparameter classifier via two routes: K-fold cross validation or Grid Search Cross Validation.  Below are the steps.
 

**Tuning model with K-fold cross validation**

1. First we use the cv_score function to perform K-fold cross-validation.  
2. Second, we use the scoring function to test each fold.
3. Third, through the now optimized cv_function, we find find best model C parameters based only on the training set, without showing the model test set at all .
4. Fourth, we create a logistic regression model with the newly calculated  best regularization parameter C and train, only again,  on the training data.
5. Fifth, finally, we train logistic regression (using the best regularization parameter C and the predictive function) on TEST data
6. Sixth, we use the accuracy_score function to test on the models accuracy for predicting y values.


**Tuning model with Scikit-leanr’s Grid Search Cross Validation**

Grid Search CV uses the best model over the training set, in comparison to K-fold Cross validation, which utilizes a iterations of loops over model parameters.

1. First we use the grid search function to apply logistic regression on the parameters using cross validation
2. Second, We fit the training set, not the test data, and find the best 
    1. Estimator C and its respective parameters 
    2. Best score for the test data and confirmation of chosen parameters
        1. mean scores of grid scores
        2. stds
        3. parameters 
3. Third, we use the best regularization estimator C found in 2.A, to fit the testing data 
4. Fourth,  we use the accuracy_score function to test on the  models accuracy for predicting y values.


Note, that for K-fold/Grid Search cross validation, a very important parameter is shared, the regularization parameter C.  Binary classification models have hyperparameters that can tuned for better performance.  In Logistic Regression, the most important parameter to tune is the regularization parameter C.  This parameter controls for unlikely high regression coefficients and can be used as a method for feature selection when data is sporadically dispersed.  In order to select the best regularization parameter C for our model, we must follow a few steps. 

We are ready to explore our data with Logistic Regresion

```python
# 1) Train/test splitting

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score



# Split the data into a training and test set.
Xlr_all, Xtestlr_all, ylr_all, ytestlr_all = train_test_split(X, y_all ,random_state=5)
```

```python
# 5) NULL ACCURACY FOR BINARY CLASSIFICATION 

# Everytime classification accuracy is used as a model evaluation metric, it's important to compare with null accuracy.
# Null acccuracy is the accuracy that could be achieved by always predicitng the most frequent class in the 
# evaluated testing/training set

# A. Calculate % of 1 values
print('% of 1 ytest values (no_CV) for all drugs: ' + str((ytestlr_all.mean())))

# B. Calculate 5 of 0 values 
print('% of 0 ytest values (no_CV) for all drugs: ' + str((1- ytestlr_all.mean())))

# C. The larger of the %'s is he Null Accuracy for this problem which is 0.997876857749
print('The Null Accuracy for this problem is: ' + str((ytestlr_all.mean())))
      

    
# Classification accuracy is the easiest classificication metric to understand but it doesnt tell us about the 
# underlying distribution of the response values or types errors the classifier is making and thus null accuracy becomes
# important.  Our null accuracy or percentage of 1 ytest values (no_CV) for all drugs is 0.997876857749 or rather, 
# the percenatge our model predicts whether a particpant will use drugs or not, will be correct about 0.9978% of the 
# time.
```

```python
# 8) ROC Curve

# Examination of plotting the ROC Curve to identify how sensitivity and specificity are affected by various 
# thresholds without actually changing the threshold.

# Firs lets evaluate Specificity

#   A. Adjust the classification threshold and print the predicted probabilities for class 1 

thresh_prob_all_no_CV = clf.predict_proba(Xtestlr_all)[:, 1]
thresh_prob_all_no_CV
```

```python
# The ROC curve is a plot of the true positive rate vs false positive rate for all possible classfication thresholds
# It helps us visually choose a threshold that balances sensitivity and specificity. Specificity means when the 
# actual value is negative, how often is the prediction correct. In other terms, how selective is the classifier
# in prediciting positive instances. 

#    B.  Run ROC curve function
#    C. Pass true values of testing set and predicted probabilities for class 1 for each observation

fpr_all_no_CV, tpr_all_no_CV, thresholds_all_no_CV = roc_curve(ytestlr_all, thresh_prob_all_no_CV)
plt.plot(fpr_all_no_CV, tpr_all_no_CV)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for All drugs classfier (No C.V.)')
plt.xlabel('False Postive Rate (1- Specificity)')
plt.ylabel('True Positive Recall rate (Sensitivity)')
plt.grid(True)

# ROC describes how our classifier is performing and its main advantage is that they dont require a classification 
# threshold and thus useful for high class imbalance 
```

```python
# define a fucntion that accepts a threshold and prints sensitivity and specificity 
def evaluate_threshold_all_no_CV (threshold_all_no_CV ):
    print ('Sensitivity: ', tpr_all_no_CV[thresholds_all_no_CV  > threshold_all_no_CV ][-1])
    print ('Specificity: ', 1 - fpr_all_no_CV[thresholds_all_no_CV  > threshold_all_no_CV][-1])
  
evaluate_threshold_all_no_CV(0.5)
```

```python
# 9) AUC (area under ROC curve)

# A higher AUC value is indicative of a better overall classifier and often used as a single number summary of the 
# performance of a classifier and and an alternative to the classification accuracy.  AUC can also be evaluated as a 
# representation (if you choose one random and one negative observation) of the likelihood that this classifer 
# will assign a higher predicted probability to the positive observation 


AUC_all_drugs_no_CV = roc_auc_score(ytestlr_all, y_class_predicit_all_no_CV)
print('AUC score for all drugs (no C_V) is: ' + str((AUC_all_drugs_no_CV)))
```

*Logisitc Regression Model Evaluation of K train/test splitting*

Teh system creates K train/test splits (into 2 pieces) averages results together .  This allows the model to be trained and tested independently.  It's a better estimate of out-of-shape sample performance than just the test/train splitting, however, it runs K times slower than train/test sp.

#### a. Train set

*Logisitc Regression Model Evaluation Metrics using Optimized Regularization Parameter C and K-Fold Cross Validation: TRAINING SET*

    1) K-folds Train/test splitting
    2) Tune model
    3) Set up of Logisitc Regression model using best Regularization Parameter C
    4) Class prediction 
    5) Classification Accuracy 
    6) Null Accuracy
    7) Classification Sensitivity Metric or Recall metric 
    8) Classification Precision Metric
    9) ROC 
    10) AUC (Area under the Curve)
    11) Classification Report

```python
# 1) K-folds Train/test splitting for training data

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

clf15 = LogisticRegression()

def cv_score_train_k_o(clf15, x_opi, y_opi, score_func=accuracy_score):
    result = 0
    nfold = 5
    for train, test in KFold(nfold).split(x_opi): # split data into train/test groups, 5 times
        clf15.fit(x_opi[train], y_opi[train]) # fit
        result += score_func(clf15.predict(x_opi[test]), y_opi[test]) # evaluate score function on held-out data
    return result / nfold # average


```

```python
clf16 = LogisticRegression()
CV_score_train_opi_no_C = cv_score_train_k_o(clf16, Xlr_opioids, ylr_opioids)
cv_score_train_k_o(clf16, Xlr_opioids, ylr_opioids)
print('Cross Validation precision for Opioids train data with k fold without using parameter C: ' + str((CV_score_train_opi_no_C)))
```

```python
# 2) Tune Model for Opioids k-fold CV

clf17 = LogisticRegression()
#the grid of parameters to search over

Cs = [0.001, 0.1, 1, 10, 100]
from sklearn.linear_model import LogisticRegression
highest_score = 0

for C in Cs:
        clf17 = LogisticRegression(C=C)
        avg_score = cv_score_train_k_o(clf17, Xlr_opioids, ylr_opioids)

        if avg_score > highest_score:
            highest_score = avg_score
            C3=C
  
print('Highest score parameter using the CV_function for Opioids drugs of train data: ' + str((highest_score)))
print('')
print('Best Regularization Parameter C: ' + str((C3)))
```

```python
# 3) Set up of Logisitc Regression model using best Regularization Parameter C for k-fold training

clf18 =LogisticRegression(C=C3)
clf18.fit(Xlr_opioids, ylr_opioids)
Fit_train_opi_yes_CV = clf18.fit(Xlr_opioids, ylr_opioids)
print(Fit_train_opi_yes_CV)

```

```python
# 1) K-folds Train/test splitting for testing data

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

clf19 = LogisticRegression()

def cv_score_test_k_ot(clf19, x_opit, y_opit, score_func=accuracy_score):
    result = 0
    nfold = 5
    for train, test in KFold(nfold).split(x_opit): # split data into train/test groups, 5 times
        clf19.fit(x_opit[train], y_opit[train]) # fit
        result += score_func(clf19.predict(x_opit[test]), y_opit[test]) # evaluate score function on held-out data
    return result / nfold # average
```

```python
clf20 = LogisticRegression()
CV_score_test_opi_no_C = cv_score_test_k_ot(clf20, Xtestlr_opioids, ytestlr_opioids)
cv_score_test_k_ot(clf20, Xtestlr_opioids, ytestlr_opioids)
print('Cross Validation precision for Opioids test data with k fold without using parameter C: ' + str((CV_score_test_opi_no_C)))
```

```python
# 2) Tune Model for Opioids k-fold CV for testing

clf21 = LogisticRegression()
#the grid of parameters to search over

Cs = [0.001, 0.1, 1, 10, 100]
from sklearn.linear_model import LogisticRegression
highest_score = 0

for C in Cs:
        clf21 = LogisticRegression(C=C)
        avg_score = cv_score_test_k_ot(clf21, Xtestlr_opioids, ytestlr_opioids)

        if avg_score > highest_score:
            highest_score = avg_score
            C4 =C

print('Highest score parameter using the CV_function for Opioids of test data: ' + str((highest_score)))
print('')
print('Best Regularization Parameter C: ' + str((C4)))
```

```python
# 3) Set up of Logisitc Regression model using best Regularization Parameter C for k-fold testing

clf22 =LogisticRegression(C=C4)
clf22.fit(Xtestlr_opioids, ytestlr_opioids)
Fit_test_opi_yes_CV = clf22.fit(Xtestlr_opioids, ytestlr_opioids)
print(Fit_test_opi_yes_CV)
```

```python
'''Simulate many experiments and see how much the results vary 
from one experiment to the next. '''

#Returns a random sample of items from an axis of object.
def make_sample(n):
    sample = data15_2['neuroticism'].sample(n) 
    return sample
#Calculates mean of sample
def sample_stat(sample):
    return sample.mean()

def compute_sample_statistics(n, iters):
    stats = [sample_stat(make_sample(n)) for i in range(iters)]
    return np.array(stats)

#Calculates mean of 1000 random samples, each of size 100
sample_means = compute_sample_statistics(n=100, iters=1000)

plt.hist(sample_means)
plt.xlabel('sample mean')
plt.ylabel('count')
None
```

```python
# Now lets find the standard error by taking the standard deviation of the sample means 
# which quantifies the variability from one experiment to the next, and reflects the precision of the estimate.
std_err = sample_means.std()
std_err
```

```python
#We can also use the distribution of sample means to compute a "90% confidence interval", 
#which contains 90% of the experimental results:

conf_int = np.percentile(sample_means, [5, 95])
conf_int
```

```python
#Now we'd like to see what happens as we vary the sample size, n. 
#The following function takes n, runs 1000 simulated experiments, salculates the Standard Error and 90% Confdence Interval
def summarize_sampling_distribution(sample_stats):
    print('Standard Error', sample_stats.std())
    print('90% Confdence Interval', np.percentile(sample_stats, [5, 95]))

summarize_sampling_distribution(sample_means)


#Answer:  As we can see our standard error and C.I of our sample vs the randomized are very close and thus we can 
#conclude that the sample size is appropiate and independent 
```

#Question 2: Can we createa a null and alternate hypotheses?

#H0 : prob_non = prob_user
#H1 : prob_non != prob_user

The null hypothesis or H0 is that there is no difference in the mean rates between someone using legal highs or not 
not using legal highs based on their neuroticism score.  The alternate hypothesis or H1 is that there is a difference 
between someone using legal highs or not not using legal highs based on their neuroticism score. This makes our problem binomial.

When testing for the difference between two proportions, your null hypothesis is that there is NO difference between the two. The alternative hypothesis is that their difference is statistically different from zero. 

### The Hypothesis

The null hypothesis is that there is no difference between the two proportions.
The alternative is that there is a statistical difference between the two proportions.

Before this test began, we would have picked a significance level. Let's just say it's 95%. Since this is a test for the difference, it's a two-tail test so a z-score past -1.96 or 1.96 will be significant.



```python
#Question 3: Exploration of gender and Opioids Cluster 

#Compute margin of error, confidence interval, and p-value.

#A. The first thing we need to do is find the null H0 and alternative H1 values for heroin usage

prob_non2 = sum(data15[data15.legalh==0].neuroticism) 
prob_user2 = sum(data15[data15.legalh==1].neuroticism)
n_h2 = prob_non2 + prob_user2
print('Total Legal Highs population: ' + str((n_h2)))
print('H0: Total Legal High Non-users with postive neuroticism trait: ' + str((prob_non2))) 
print('H1: Total Legal High Users with postive neuroticism trait: ' + str((prob_user2))) 
```

```python
#total population= 212, H0= 66, H1= 146
##total population= 4900, H0= 3486, H1= 1414

# sample mean= X_bar= 1(H0) + 0(H1)/population = (1(66) + 0(146))/212
### sample mean= X_bar= 1(H0) + 0(H1)/population = (1(3486) + 0(1414))/4900

#variance = S^2= [H0(1-sample mean)^2 + H1(0-sample mean)^2]/population -1= 66(1- 0.40)^2+ 146(0-.40)^2/population -1
##variance = S^2= [H0(1-sample mean)^2 + H1(0-sample mean)^2]/population -1= 66(1- 0.40)^2+ 1414(0-.29)^2/4900 -1


import math as math
n_h2= 4900
X_bar2= ((1*prob_non2) + (0*prob_user2))/(n_h2)
print('Sample Mean ' + str(( X_bar2)))

STD2= math.sqrt(((3486*(1-X_bar2)**2) + (prob_user2*(0-X_bar2)**2))/(n_h2-1))
print('Standard Deviation= ' + str(( STD2)))

Ox_2= STD2/math.sqrt(n_h2)
Ox_perc2= Ox_2*100
print('Distribution of the sample mean= ' + str(( Ox_perc2)) + '% ')

#According to the Z score table, the margin of error with a STD at 0.4531 is 1.49 +- 0.013(2* OX_2) = 1.503 & 1.477
print('Z score= ' + str((1.49)) + ' +/- 0.013' )

#Margin of Error= Zscore * Ox_2(Distribution of the sample mean)
ME=1.49 * Ox_2
ME_perc =ME*100
print('Margin of Error= ' + str((ME_perc)) + '%')

#95% Confidence Interval
#=Sample Mean % +-  Distribution of the sample mean %
#=Sample Mean + 
print('95% Confidence Interval of 71.78% and 70.49%')


from statsmodels.stats.proportion import proportions_ztest as pz
p_values2 = pz(np.array([1320, 564]),np.array([3486, 1414]),value=0)
print('Z score, P value ' + str(( p_values2)))
```
