```python
#Question 1: What test is appropriate for this problem? Does CLT apply?
#lets examine and define how many characters are in the gender category 
data3["neuroticism"] = data3[["neuroticism"]].astype(int)
df_male = data3[data3.gender=='male']
df_female = data3[data3.gender=='female']
num_male = len(df_male)
num_female = len(df_female)
print('Total population of samples ' + str(len(data3.gender))) 
print('"male" race assigned randomly ' + str(num_male))
print('"female" race are assigned randomly ' + str(num_female))

#The first thing we look for to determine if the distribution is normal it to examine if the total
# population (n) is >30. Based on previous cell, our n = 1884, 943 for male, 941 for female.  Secondly, 
# we need to determine what type of problem this is.  It is a binomial problem as random 
# candidates either took a drug or did not take a drug.  Due to the size of samples, a binomial distribution can be 
# approximated by a normal distribution per the CLT.
```

```python
#Question 2: What are the null and alternate hypotheses?

#H1 : prob_m = prob_f
#H0 : prob_m != prob_f

#The null hypothesis or H0 is that there is no difference in the mean rates between someone taking a drug or not taking
# a drug.  The alternate hypothesis or H1 is that there is a difference between someone taking a drug and someone 
# not taking a drug
```

```python
#Question 3: Compute margin of error, confidence interval, and p-value.

#A. The first thing we need to do is find the null H0 and alternative H1 values for heroin usage
m_heroin = sum(data3[data3.gender=='male'].heroin) 
f_heroin = sum(data3[data3.gender=='female'].heroin)
n_h = m_heroin + f_heroin
print('A.total population using heroin ' + str(( f_heroin + m_heroin)))
print('  H0: total male using heroin  ' + str((m_heroin))) 
print('  H1: total female using heroin ' + str((f_heroin))) 


#A.ii. The first thing we need to do is find the nullH0 and alternative H1 values for legalh usage
m_legalh = sum(data3[data3.gender=='male'].legalh) 
f_legalh = sum(data3[data3.gender=='female'].legalh)
n_h= m_legalh + f_legalh
print('B.total population using legalh ' + str(( f_legalh + m_legalh)))
print('  H0: total male using legalh  ' + str((m_legalh))) 
print('  H1: total female using legalh ' + str((f_legalh))) 



#A. The first thing we need to do is find the null H0 and alternative H1 values for meth usage
m_meth = sum(data3[data3.gender=='male'].meth) 
f_meth = sum(data3[data3.gender=='female'].meth)
n_m= m_meth + f_meth
print('C.total population using meth ' + str(( f_meth + m_meth)))
print('  H0: total male using meth  ' + str((m_meth))) 
print('  H1: total female using meth ' + str((f_meth))) 

```

```python
#total population= 212, H0= 66, H1= 146
# sample mean= X_bar= 1(H0) + 0(H1)/population = (1(66) + 0(146))/212
#variance = S^2= [H0(1-sample mean)^2 + H1(0-sample mean)^2]/population -1= 66(1- 0.40)^2+ 146(0-.40)^2/population -1

import math as math
n_h= 392
X_bar= ((1*f_heroin) + (0*m_heroin))/(n_h)
print('Sample Mean ' + str(( X_bar)))

STD= math.sqrt(((f_heroin*(1-X_bar)**2) + (m_heroin*(0-X_bar)**2))/(n_h-1))
print('Standard Deviation ' + str(( STD)))

Ox_= STD/math.sqrt(n_h)
Ox_perc= Ox_*100
print('Distribution of the sample mean ' + str(( Ox_)) + ' or % ' + str(( Ox_perc)))

#According to the Z score table, the margin of error with a STD at 0.3568 is 1.07 +- 0.1683 or 1.24 and 0.90
#the margin of error at 95% interval is 

print('Margin of error =', 1.96 * STD)
print('Range for 95% confidence interval:', (1.24 * STD, 0.90 * STD))

##Because our CI is <5 we keep the null value and accept it 

from statsmodels.stats.proportion import proportions_ztest as pz
p_values = pz(np.array([f_heroin,m_heroin]),np.array([num_female,num_male]),value=0)
print('P_values ' + str(( p_values)))
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

def cv_score(clf, x, y, score_func=accuracy_score):
    result = 0
    nfold = 5
    for train, test in KFold(nfold).split(x): # split data into train/test groups, 5 times
        clf.fit(x[train], y[train]) # fit
        result += score_func(clf.predict(x[test]), y[test]) # evaluate score function on held-out data
    return result / nfold # average

```

```python
clf = LogisticRegression()
score_all_drugs = cv_score(clf, Xlr_all, ylr_all)
print(score_all_drugs)
```

```python
#the grid of parameters to search over
Cs = [0.001, 0.1, 1, 10, 100]
from sklearn.linear_model import LogisticRegression
highest_score = 0

for C in Cs:
        clf = LogisticRegression(C=C)
        avg_score = cv_score(clf, Xlr_all, ylr_all)

        if avg_score > highest_score:
            highest_score = avg_score
            C1 =C
print (highest_score, C1)
```

```python
##By using the best C found in the previous example, we train logistic regression on the training data
##and then we use that on the test data for X by using the predict function.  To test its accuracy we then use the
##accuracy_score fucntion by predicting the y values
clfl=LogisticRegression(C=C1)
clfl.fit(Xlr_all, ylr_all)
ypred=clfl.predict(Xtestlr_all)
accuracy = accuracy_score(ypred, ytestlr_all)
print(accuracy)
```
